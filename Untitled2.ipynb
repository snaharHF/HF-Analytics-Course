```python
import numpy as np
import pandas as pd
import seaborn as sns
url = 'https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv'
df = pd.read_csv(url)
print(df.head())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
letters_only = df['School DBN'].str.replace(r'[^a-zA-Z]', '', regex=True)

print(letters_only)
```

```python
drop_cols = ['Number Scoring Below 65', 'Number Scoring 65 or Above', 'Number Scoring 80 or Above', 'Number Scoring CR', 'Percent Scoring CR']
df.drop(drop_cols, inplace = True, axis = 1)
```

```python
print(df.head())
```

```python
df['Regents Exam'].fillna('Not Available', inplace = True)
df.isnull().sum()
```

```python
### Replacing all the 's' values to NAN so we can change the column type to a float.

df['Mean Score'] = df['Mean Score'].replace('s', np.nan)
df['Mean Score'] = df['Mean Score'].astype(float)

df['Percent Scoring Below 65'] = df['Percent Scoring Below 65'].replace('s', np.nan)
df['Percent Scoring Below 65'] = df['Percent Scoring Below 65'].astype(float)

df['Percent Scoring 65 or Above'] = df['Percent Scoring 65 or Above'].replace('s', np.nan)
df['Percent Scoring 65 or Above'] = df['Percent Scoring 65 or Above'].astype(float)

df['Percent Scoring 80 or Above'] = df['Percent Scoring 80 or Above'].replace('s', np.nan)
df['Percent Scoring 80 or Above'] = df['Percent Scoring 80 or Above'].astype(float)
```

```python
### Focusing on K-8 Common Core Algebra exam in 2016.

df = df.drop(df[df['Year'] != 2016].index)
df = df.drop(df[df['School Level'] != 'K-8'].index)
df = df.drop(df[df['Regents Exam'] != 'Common Core Algebra'].index)
```

```python
### Breaking out the schools into boroughs, mainly focusing in Manhattan and the Bronx.

df['Borough'] = np.where(df['School DBN'].str.contains('M'), 'Manhattan',
                         np.where(df['School DBN'].str.contains('X'), 'Bronx', 'Other'))

print(df.head())
```
